```python
import matplotlib.pyplot as plt
csvFile = pd.read_csv('FINAL different reactions glu vs lac 4 decimals -30 to 30.csv')

r_id= csvFile.iloc[0,3]
csvFile2= pd.read_csv('biomass_per_flux_for_reaction_{}.csv'.format(r_id))
csvFile2

data = {'flux': csvFile2.iloc[:,1],
        'Biomass': csvFile2.iloc[:,2]
       }

df = pd.DataFrame(data)

plt.plot(df['flux'], df['Biomass'], marker='o')
plt.title('Biomass per flux for reaction {}'.format(r_id))
plt.xlabel('flux')
plt.ylabel('Biomass')

from matplotlib import pyplot as plt
plt.savefig('image 4 decimals.png')

plt.show()
```

```python
from PIL import Image

file_in = "image 4 decimals.png"
img = Image.open(file_in)
file_out = 'test1.bmp'
print (len(img.split())) # test
if len(img.split()) == 4:
    # prevent IOError: cannot write mode RGBA as BMP
    r, g, b, a = img.split()
    img = Image.merge("RGB", (r, g, b))
    img.save(file_out)
else:
    img.save(file_out)

from xlwt import Workbook
w = Workbook()
ws = w.add_sheet('Images')
ws.insert_bitmap(file_out, 0, 0)
w.save('images 4 decimals -30 to 30.xls')
```
